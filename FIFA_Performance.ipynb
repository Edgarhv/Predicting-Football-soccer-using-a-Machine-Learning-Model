```python
import pandas as pd
fifa18_df = pd.read_csv("fifa18_clean.csv")
fifa18_df = fifa18_df[["Name", "Age", "Overall", "Acceleration", "Agility", "Balance", "Ball control", "Dribbling", "Free kick accuracy", "Strength", "Stamina", "Sprint speed", "Shot power"]]
fifa18_df['season'] = 2018
fifa18_df
```

```python
fifa18_df.rename(columns={"Name": "player_name",
                   'Age': 'age',
                   'Acceleration':'acceleration',
                   'Overall':'overall',
                   'Agility':'agility',
                   'Balance':'balance',
                   'Ball control':'ball_control',
                   'Dribbling':'dribbling',
                   'Free kick accuracy':'free_kick_accuracy',
                   'Strength':'strength',
                   'Stamina':'stamina',
                   'Sprint speed':'sprint_speed',
                   'Shot power':'shot_power'}, inplace=True)
fifa18_df
```

```python
import plotly.graph_objects as go
import plotly.express as px
import plotly
attributes = ['Acceleration','Shooting','Agility','Dribbling','Speed','Strength','Stamina','Overall']
def player_performance(name):
    data = fifa18_df[fifa18_df.player_name.str.startswith(name)]
    trace = go.Scatterpolar(
    r= [data['acceleration'].values[0],data['shot_power'].values[0],
              data['agility'].values[0],data['dribbling'].values[0],
              data['sprint_speed'].values[0],data['strength'].values[0],
              data['stamina'].values[0],data['overall'].values[0]
              ],
           theta =  attributes,
           fill  = 'toself',
           name  = '2018'
    )
    data_trace=[trace]
    layout = go.Layout(
                    polar=dict(
                           radialaxis = dict(
                                         visible=True,
                                         range=[0,100]
                           )
                    )
        ,showlegend=True,
        title='Stats Related to {} From 2018'.format(name)
    
    )
    fig=go.Figure(data=data_trace,layout=layout)
    fig.show()
```

```python
player_performance('Cristiano Ronaldo')
```
